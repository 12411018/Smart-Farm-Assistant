## Load All Documents Using LangChain

The `load_all_documents()` function from `data_loader.py` uses LangChain loaders to extract text from all supported file types in the data directory.

```python
# Load all documents from the data directory using data_loader.load_all_documents()
# This scans the data folder for all supported file types recursively
documents = load_all_documents("data")

print(f"\n✓ Successfully loaded {len(documents)} total documents")
print(f"\nDocuments loaded from:")
print(f"  - PDF files (PyPDFLoader)")
print(f"  - TXT files (TextLoader)")
print(f"  - CSV files (CSVLoader)")
print(f"  - Excel files (UnstructuredExcelLoader)")
print(f"  - Word files (Docx2txtLoader)")
print(f"  - JSON files (JSONLoader)")
```

## Display Loaded Documents with LangChain Metadata

LangChain documents contain both content and metadata. View details of loaded documents:

```python
if documents:
    print(f"Total documents loaded via LangChain: {len(documents)}\n")
    print("First 6 LangChain documents:\n")
    for i, doc in enumerate(documents[:6], 1):
        print(f"Document {i}:")
        print(f"  Source file: {doc.metadata.get('source', 'Unknown')}")
        print(f"  Content length: {len(doc.page_content)} characters")
        print(f"  LangChain metadata: {doc.metadata}")
        print(f"  Content preview: {doc.page_content[:100]}...")
        print()
else:
    print("No documents found. Please add files to the 'data' directory.")
```

## Document Summary Statistics

Summary of loaded documents by type and count.

```python
from collections import defaultdict

# Categorize documents by source file type
file_types = defaultdict(int)
total_chars = 0

for doc in documents:
    source = doc.metadata.get('source', 'Unknown')
    file_ext = source.split('.')[-1].lower() if '.' in source else 'unknown'
    file_types[file_ext] += 1
    total_chars += len(doc.page_content)

print("Document Statistics:")
print(f"  Total documents: {len(documents)}")
print(f"  Total characters: {total_chars:,}")
print(f"\nDocuments by file type:")
for file_type, count in sorted(file_types.items()):
    print(f"  .{file_type}: {count} document(s)")
```
